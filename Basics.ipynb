I am making the category for the different experience people
New: Less than 3 years at a company
Experienced: 3-6 years at a company
Established: 7-10 years at a company
Veteran: 11 or more years at a company

```python
def career_stage(val):
    if val <3:
        return 'New'
    elif val>=3 and val <=6:
        return 'Experienced'
    elif val>=7 and val<=10:
        return 'Established'
    elif val >=11:
        return 'Veteran'
    elif pd.isnull(val):
        return np.nan
        
        
```

```python
combined_update['service_cat']=combined_update['institute_service_up'].apply(career_stage)
```
